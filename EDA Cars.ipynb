```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
from matplotlib import pyplot as plt
```

```python
cars_data = pd.read_csv('D:/Galang/mtcars.csv')
cars_data.head()
```

```python
def new(cars_data):
    if  cars_data['mpg'] <= 20 :
        result = 'low'
    elif 20 <= cars_data['mpg'] <= 30 :
        result = 'medium'
    else:
        result = 'high'
    return result
```

```python
cars_data['mpg_level'] = cars_data.apply(new, axis=1)
```

```python
cars_data.head()
```

```python
cars_data.describe()
```

```python
cars_data.describe(include='all')
```

```python
print(cars_data[cars_data['mpg_level'] == 'low']['mpg'].count())
print(cars_data[cars_data['mpg_level'] == 'medium']['mpg'].count())
print(cars_data[cars_data['mpg_level'] == 'high']['mpg'].count())
```

```python
cars_data['mpg_level'].value_counts().plot.bar(title='Freq dist of mpg_level')
```

```python
cars_data_low = cars_data[cars_data['mpg_level'] == 'low']
cars_data_med = cars_data[cars_data['mpg_level'] == 'medium']
cars_data_high = cars_data[cars_data['mpg_level'] == 'high']
```

```python
col_names = ['mpg','cyl', 'disp', 'hp', 'wt', 'drat']

fig, ax = plt.subplots(len(col_names), figsize=(16,12))

for i, col_val in enumerate(col_names):

    sns.distplot(cars_data_low[col_val], hist=True, ax=ax[i])
    ax[i].set_title('Freq dist '+col_val, fontsize=10)
    ax[i].set_xlabel(col_val, fontsize=8)
    ax[i].set_ylabel('Count', fontsize=8)

plt.show()
```

```python
cars_num = cars_data.select_dtypes(include = ['int64','float64'])
cars_num.head()
```

```python
cars_num.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8)
```

```python
cars_num_corr = cars_num.corr()['mpg'][:-1]
golden_features_list = cars_num_corr[abs(cars_num_corr) > 0.5].sort_values(ascending=False)
print("There is {} strongly correlated values with mpg:\n{}".format(len(golden_features_list), golden_features_list))
```

```python
for i in range(0, len(cars_num.columns), 5):
    sns.pairplot(data=cars_num,
                x_vars=cars_num.columns[i:i+5],
                y_vars=['mpg'])
```

```python
corr = cars_num.drop('mpg', axis=1).corr() 
plt.figure(figsize=(12, 10))

sns.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)], 
            cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 8}, square=True);
```
